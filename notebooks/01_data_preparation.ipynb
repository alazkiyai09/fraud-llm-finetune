```python
import json
from pathlib import Path

for split in ['train', 'val', 'test']:
    path = Path('../data/processed') / f'{split}.jsonl'
    rows = [json.loads(line) for line in path.read_text(encoding='utf-8').splitlines() if line.strip()]
    print(split, len(rows))
```
